# **RIDGE REGRESSOR**

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
parameters = {'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40, 45, 50, 55, 100]}
ridge_regressor = GridSearchCV(ridge, parameters, scoring='neg_mean_squared_error', cv=5)
ridge_regressor.fit(x, y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
pred_ridge = ridge_regressor.predict(x_test)
```

```python
pred_ridge
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

MAE = mean_absolute_error(y_test,pred_ridge)
print(MAE)

MSE = mean_squared_error(y_test,pred_ridge)
print(MSE)

RMSE = np.sqrt(MSE)
print(RMSE)

lrmse = np.log(RMSE)
print(lrmse)
```

# **GridSearchCV**


```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import RepeatedKFold
```

```python
from sklearn.linear_model import ElasticNet
```

```python
model = ElasticNet()
```

```python
em = RepeatedKFold(n_splits = 10, n_repeats = 5, random_state = 1)
```

```python
grid = dict()
grid ['alpha'] = [1e-3,1e-2,1e-1,0.0,1.0,2.0,3.0]
```

```python
grid['l1_ratio'] = np.arange(0,1,0.01)
```

```python
search = GridSearchCV(model, grid, scoring = 'neg_mean_absolute_error', cv = em, n_jobs = -1)
```

```python
results = search.fit(x,y)
```

```python
print('MAE:%3.f' % results.best_score_)
```

```python
print("best alpha :",results.best_params_)
```
